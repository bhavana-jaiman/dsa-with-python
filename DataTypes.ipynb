# Built-in Data Types & Data Structures

### 1.1 Numeric Types

```python
# numeric types
a = 5
type(a)
```

```python
# float
b = 3.14
c = 24.
type(b)
type(c)
```

```python
# complex
d = 2 + 3j
type(d)
```

### 1.2 Sequence Type

```python
# str -> Text(string of characters)
a = "Hello, World!"
type(a)
```

```python
b = 'Python is great!'
type(b)
```

```python
c = '''This is a multi-line string.'''    # multi line string
type(c)
```

```python
# list -> Mutable ordered collection of items
L = [1, 2, 3, 4, 5.1, "Hello", [6, 7], (8, 9), {10: "ten"}, {11, 12}]  #any num of elements and any type of elements
type(L)
```

```python
# tuple -> Immutable ordered collection of items
T = (1, 2, 3, 4, 5.1, "Hello", [6, 7], (8, 9), {10: "ten"}, {11, 12})  #any num of elements and any type of elements
type(T)       # comma makes it tuple, even without parentheses
```

```python
# range -> Immutable sequence of numbers
r = range(10)  # 0 to 9
type(r)
# print(list(r))  # convert range to list to see the numbers
```

### 1.3 Set Types

```python
# set -> mutable, unordered, unique items
s = {1, 2, 3, 4, 5, 5, 6}
type(s)
# s
```

```python
# frozenset -> immutable, unordered, unique items
fs = frozenset({1, 2, 3, 4, 5, 5, 6})
type(fs)
```

### 1.4 Mapping Types

```python
# dictionary -> mutable, unordered collection of key-value pairs
d = {"name": "Alice", "age": 30, "city": "New York"}
type(d)
```

### 1.5 Boolian Types

```python
# Bool -> True or False
b1 = True
b2 = False
type(b1)
type(b2)
```

### 1.6 None Type

```python
a = None
type(a)
```
