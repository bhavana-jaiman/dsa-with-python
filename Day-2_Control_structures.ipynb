### 2.1 if-else

```python
a = 7 
if a > 5:
    print("a is greater than 5")
else:
    print("Not greater than 5")
```

```python
a = 5
if a %2 == 0:
    print("Even")
else:
    print("Odd")
```

### 2.2 if-elif-else

```python
marks = 74

if marks > 90:
    print("Excellent")
elif marks > 80:
    print("Very Good")
elif marks > 70:
    print("Good")
else:
    print("Poor")
```
