# 3 Loops

### 3.1 For Loop

### if we know how many iterations are there

```python
L1 = [1,2,3,6,9]
print(L1[0])  # Output: 1
print(L1[1])  # Output: 2
print(L1[2])  # Output: 3
print(L1[3])  # Output: 6
print(L1[4])  # Output: 9
     
```

```python
L1 = [1,2,3,6,9]
for i in L1:
    print(i)
```

```python
for i in range(20,31):
    if i % 2 == 0:
        print(i)
```

```python
for i in range(20,31, 2):         # without check
    print(i)
```

### 3.2 While Loop

### we don't know the iterations but conditional

```python
# condition based statements
a = 1
while a <= 5:
    print(a)
    a += 1
   
```
