# 4 Functions

### 4.1 Define/Call Function

```python
def random_func():
    print("This is a random function")
```

```python
random_func()
```

```python
def add_number():
    a = 2+3
    print(a)
```

```python
add_number()
```

### 4.2 Input to Function

```python
def add_two_numbers(a,b):           # parameters
    print(f"sum of {a} and {b} is {a+b}")
```

```python
add_two_numbers(2,3)    # arguments
```

```python
def greeting(name):
    print(f"Hello {name}")
```

```python
greeting("Bhavana")
```

### 4.3 Output from Function

```python
def add_two_numbers(a,b):           
    result = a+b
    print(f"sum of {a} and {b} is {a+b}")
    return result
```

```python
c = add_two_numbers(2,3)
```

```python
print(c)
```
