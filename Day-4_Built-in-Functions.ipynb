### 5.7 enumerate

```python
names = ['aaron','ben','charlie','david','edward','aron']
```

```python
for i in range(len(names)):
    print(i, names[i])
```

rather than we can do

```python
for x in enumerate(names):
    print(x)
```

```python
for x in enumerate(names,start=1):
    print(x)
```

### 5.8 zip

need to iterate over multiple sequences at same time

```python
names
```

```python
colors = ['red','green','blue','yellow','orange','purple']
age = [23,45,12,34,56,78]
```

```python
for x in zip(names, age, colors):
    print(f'{x[0]} is {x[1]} years old and likes {x[2]} color')
```
