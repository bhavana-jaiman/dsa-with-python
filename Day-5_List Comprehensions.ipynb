# 6 List Comprehenion

```python
result = []

for i in range(1,11):
    if i%2 == 0:
        result.append(i)

result
```

```python
list(range(2,11,2))
```

```python
# List comprehension
seq = range(1,11)
[a for a in seq if a%2 ==0]
```

```python
ages = [23,45,12,34,56,78]

['Greater than 30 'if age > 30 else 'less than 30' for age in ages]
```
