```python
class MyRectangle:
    def __init__(self,length,breadth):      #attribute is property of the class
        self.length = length
        self.breadth = breadth
        self.random_attribute = 0   # we don't need to pass the attribute by user we can make it's own 
    def get_perimeter(self):
        return 2*(self.length + self.breadth)    # Method    
    def get_area(self):
        return self.length * self.breadth # Method
    def get_length(self):
        return self.length         # Method
```

Method is a function defines in a class
