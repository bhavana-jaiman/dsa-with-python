# 6 Array

An array is a data structure that stores a collection of elements of the same data type in contiguous memory location

# 6.1 Properties of an Array:

1. Homogeneous Elements: All elements are of the same data type(model weights of type float 32)

2. Contiguous Memory Allocation: Elements are stored one after another in RAM

3. Fixed Index Access: Can instantly fetch any element with it's index

4. Efficient Iteration: Travelling all elements is sequential and cache-friendly

# 6.2 Fixed-Size Arrays:

* Size is decided at creation

* Stored contiguously in memory

* More memory-efficient, but less flexible

* Implementation: array, numpy

* Usage: Best for predictable, stable data (offline learning)

# 6.3 Dynamic-Size Arrays:

* Can grow or shrink during runtime

* Allocates extra “spare” memory to allow future growth

* Implementation: lists

* Usage: Best for unpredictable, growing datasets (online learning)
